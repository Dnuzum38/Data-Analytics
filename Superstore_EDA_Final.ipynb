```python
# prompt: using matplotlib and seaborn display sales over time using the order date and view by month year quarter and those are not in the data set

import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'Order Date' is a datetime column
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Create new columns for month, year, and quarter
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
df['Quarter'] = df['Order Date'].dt.quarter

# Group by year and calculate total sales
sales_by_year = df.groupby('Year')['Sales'].sum()

# Group by month and year and calculate total sales
sales_by_month_year = df.groupby(['Year', 'Month'])['Sales'].sum()

# Group by quarter and year and calculate total sales
sales_by_quarter_year = df.groupby(['Year', 'Quarter'])['Sales'].sum
# Plotting sales over time

# Yearly Sales
plt.figure(figsize=(10, 6))
plt.plot(sales_by_year.index, sales_by_year.values, marker='o')
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()

# Monthly Sales
plt.figure(figsize=(12, 6))
sales_by_month_year.unstack().plot(marker='o')
plt.title('Total Sales by Month and Year')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend(title='Year')
plt.grid(True)
plt.show()

# Quarterly Sales
plt.figure(figsize=(10, 6))
sales_by_quarter_year.unstack().plot(marker='o')
plt.title('Total Sales by Quarter and Year')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.legend(title='Year')
plt.grid(True)
plt.show()
```

## time series using resample

```python
# prompt: rewrite below code using resample

import pandas as pd
import numpy as np
from google.colab import drive
import seaborn as sns
import matplotlib.pyplot as plt

# Mount Google Drive
drive.mount('/content/drive/')

# Load the dataset
file_path = '/content/drive/MyDrive/Descriptive Analytics/Olympics_EDA/Data/Global_Superstore_Orders_2012_2015.xlsx'
df = pd.read_excel(file_path, dtype={'Postal Code': str})

# Data Cleaning (Postal Code)
df['Postal Code'] = df['Postal Code'].str.zfill(5).fillna('na')


# Time Series Analysis with resample
df['Order Date'] = pd.to_datetime(df['Order Date'])
df = df.set_index('Order Date')

# Resample to yearly, monthly and quarterly frequency
yearly_sales = df['Sales'].resample('Y').sum()
monthly_sales = df['Sales'].resample('M').sum()
quarterly_sales = df['Sales'].resample('Q').sum()


# Plotting
plt.figure(figsize=(10, 6))
plt.plot(yearly_sales.index, yearly_sales.values, marker='o')
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(quarterly_sales.index, quarterly_sales.values, marker='o')
plt.title('Total Sales by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```
