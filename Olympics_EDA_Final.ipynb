# Clean Version - EDA Final


```python

# Give alias to the two data sources for the Olypics file
path1 = '/content/drive/MyDrive/Descriptive Analytics/Olympics_EDA/Data/athlete_events.csv'
path2 = '/content/drive/MyDrive/Descriptive Analytics/Olympics_EDA/Data/noc_regions.csv'

#Import pands, numpy, seaeborn, matpltlib adn connects to the google drive
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)

from google.colab import drive
drive.mount('/content/drive/')

# creates the 2 dataframes

df = pd.read_csv(path1)
df_noc = pd.read_csv(path2)

df = pd.read_csv(path1)
df_noc = pd.read_csv(path2)

#merges the 2 dataframes ans ensures they arte pulling over corretly
df = df.merge(df_noc, how="left", on="NOC")
df.head()
```

```python
df.info()
```

```python
# Convert NaN values in the medal comlumn to say no medal and the blank notes feilds to say No Comments
df['Medal'] = df['Medal'].fillna('No Medal')
df['notes'] = df['notes'].fillna('No Comments')
df.info()
```

```python


# Fill NaN values in 'Height' and 'Weight' columns with the calculated averages
df['Height'].fillna(0)
df['Weight'].fillna(0)
```

```python
# Replace Nan values in region to Unknown
df['region'] = df['region'].fillna('Unknown')
df.head(10)
```

```python
# remove rows with no values
df = df.dropna(axis=1)
df.info()
```

```python
# prompt: histogram of gold silver and bronze medal counts for top 10 teams sort most to least with a legend of gold silver and bronze with the color gold for gold with the color silver for silver and with the color bronze for bronze

import matplotlib.pyplot as plt
# Group by region and medal type, then count medals
medal_counts = df.groupby(['region', 'Medal'])['Medal'].count().unstack(fill_value=0)

# Sort by total medals (Gold + Silver + Bronze) in descending order
medal_counts['Total'] = medal_counts['Gold'] + medal_counts['Silver'] + medal_counts['Bronze']
top_10_teams = medal_counts.sort_values(by='Total', ascending=False).head(10)

# Plotting the histogram
top_10_teams[['Gold', 'Silver', 'Bronze']].plot(kind='bar', stacked=True, figsize=(12, 6), color=['gold', 'silver', 'brown'])

plt.title('Top 10 Teams by Medal Count')
plt.xlabel('Team')
plt.ylabel('Number of Medals')
plt.xticks(rotation=45)
plt.legend(['Gold', 'Silver', 'Bronze'], loc='upper right') # added legend with specified colors
plt.tight_layout() # added to prevent labels from overlapping

plt.show()
```

```python
# prompt: line chart comparing medal counts for Team United States team Russia and Team China over time fort the summer season

import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame from the previous code
# Filter for Summer Olympics
summer_df = df[df['Season'] == 'Summer']

# Select the three teams
teams = ['United States', 'Russia', 'China']
team_df = summer_df[summer_df['Team'].isin(teams)]

# Group by year and team, count medals
medal_counts_by_year = team_df.groupby(['Year', 'Team'])['Medal'].count().unstack(fill_value=0)

# Plotting
plt.figure(figsize=(12, 6))
for team in teams:
  plt.plot(medal_counts_by_year.index, medal_counts_by_year[team], label=team)

plt.title('Medal Counts for Top 3 Teams (Summer Olympics)')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# prompt: line chart comparing medal counts for Team United States team Russia and Team China over time for the summer season since 1990

import matplotlib.pyplot as plt
# Filter for Summer Olympics since 1990
summer_df_since_1990 = summer_df[summer_df['Year'] >= 1990]

# Select the three teams
teams = ['United States', 'Russia', 'China']
team_df_since_1990 = summer_df_since_1990[summer_df_since_1990['Team'].isin(teams)]

# Group by year and team, count medals
medal_counts_by_year = team_df_since_1990.groupby(['Year', 'Team'])['Medal'].count().unstack(fill_value=0)

# Plotting
plt.figure(figsize=(12, 6))
for team in teams:
    plt.plot(medal_counts_by_year.index, medal_counts_by_year[team], label=team, marker='o')  # Added markers

plt.title('Medal Counts for USA, Russia, and China (Summer Olympics since 1990)')
plt.xlabel('Year')
plt.ylabel('Number of Medals')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# prompt: number of female athletes in the summer olympics over time

import matplotlib.pyplot as plt
# Assuming 'df' is your DataFrame from the previous code
# Filter for Summer Olympics and female athletes
female_athletes_df = df[(df['Season'] == 'Summer') & (df['Sex'] == 'F')]

# Group by year and count the number of female athletes
female_athletes_by_year = female_athletes_df.groupby('Year')['ID'].count()

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(female_athletes_by_year.index, female_athletes_by_year.values, marker='o')

plt.title('Number of Female Athletes in Summer Olympics Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Female Athletes')
plt.grid(True)
plt.show()
```

```python
# prompt: number of male athletes in the summer Olympics over time

import matplotlib.pyplot as plt
# Assuming 'df' is your DataFrame from the previous code
# Filter for Summer Olympics and male athletes
male_athletes_df = df[(df['Season'] == 'Summer') & (df['Sex'] == 'M')]

# Group by year and count the number of male athletes
male_athletes_by_year = male_athletes_df.groupby('Year')['ID'].count()

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(male_athletes_by_year.index, male_athletes_by_year.values, marker='o')

plt.title('Number of Male Athletes in Summer Olympics Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Male Athletes')
plt.grid(True)
plt.show()
```
